```python
import rasterio as rio
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import rasterio.plot
import os
from datetime import datetime as dt
import matplotlib.dates as mdates

%matplotlib inline


import warnings
warnings.filterwarnings('ignore')
```

```python
# Load data
area_namerica = pd.read_csv(r'C:\Users\vicki\gee-globalsnow\class_area_namerica.csv')
area_europe = pd.read_csv(r'C:\Users\vicki\gee-globalsnow\class_area_europe.csv')
area_casia = pd.read_csv(r'C:\Users\vicki\gee-globalsnow\class_area_casia.csv')

mypath = os.path.abspath(r'C:\Users\vicki\gee-globalsnow') 
```

```python
# Plot north america area classifications
# Dataframe is from Jan-Apr, Nov-Dec
x1 = area_namerica.month
y1 = area_namerica.area_no
y2 = area_namerica.area_yes
y3 = area_namerica.area_maybe
```

```python
# Plot scatter plots as test
# Dataframe is from Jan-Apr, Nov-Dec
plt.scatter(x1, y1, c='k', marker='o', label='classified-no')
plt.scatter(x1, y2, c='b', marker='o', label='classified-yes')
plt.scatter(x1, y3, c='g', marker='o', label='classified-maybe')
plt.show()
```

```python
# Plot as water year (Nov-Apr)
y1.plot(style='o')
y2.plot(style='o')
y3.plot(style='o')
```

```python
# Plot north america classifications by water year

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(area_namerica.index.values,
        area_namerica['area_no'].values,
        '-o',color='red',linewidth=3, markersize=10)
ax.plot(area_namerica.index.values,
        area_namerica['area_yes'].values,
        '-o',color='green',linewidth=3, markersize=10)
ax.plot(area_namerica.index.values,
        area_namerica['area_maybe'].values,
        '-o', color='orange',linewidth=3, markersize=10)
ax.legend(loc='upper right')
ax.set(xlabel="Month", ylabel="Area (sq km)",
       title="Classified Area - North America")

N = 6
ind = np.arange(N)  
plt.xticks(ind, ('Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar'))
plt.legend((area_namerica['area_no'], area_namerica['area_yes'],area_namerica['area_maybe']), ('No', 'Yes', 'Maybe'))

plt.show()
plt.tight_layout()
plt.savefig('classified_area_north_america.png',transparent=True)

fig.savefig(os.path.join(mypath,'classified_area_north_america.png'))
```

```python
# Plot europe classifications by water year

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(area_europe.index.values,
        area_europe['area_no'].values,
        '-o', color='red',linewidth=3, markersize=10)
ax.plot(area_europe.index.values,
        area_europe['area_yes'].values,
        '-o', color='green',linewidth=3, markersize=10)
ax.plot(area_europe.index.values,
        area_europe['area_maybe'].values,
        '-o', color='orange',linewidth=3, markersize=10)
ax.legend(loc='upper right')
ax.set(xlabel="Month", ylabel="Area (sq km)",
       title="Classified Area - Europe")
ax.set_ylim((0,1.4e10))

N = 6
ind = np.arange(N)  
plt.xticks(ind, ('Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar'))
plt.legend((area_namerica['area_no'], area_namerica['area_yes'],area_namerica['area_maybe']), ('No', 'Yes', 'Maybe'))

plt.show()
plt.tight_layout()
plt.savefig('classified_area_europe.png',dpi=300)

fig.savefig(os.path.join(mypath,'classified_area_europe.png'))
```

```python
# Plot central asia classifications by water year

fig, ax = plt.subplots(figsize=(12,5))
ax.plot(area_casia.index.values,
        area_casia['area_no'].values,
        '-o', color='red', linewidth=3, markersize=10)
ax.plot(area_casia.index.values,
        area_casia['area_yes'].values,
        '-o', color='green', linewidth=3, markersize=10)
ax.plot(area_casia.index.values,
        area_casia['area_maybe'].values,
        '-o', color='orange', linewidth=3, markersize=10)
ax.legend(loc='upper right')
ax.set(xlabel="Month", ylabel="Area (sq km)",
       title="Classified Area - Central Asia")
ax.set_ylim((0,1.4e10))

N = 6
ind = np.arange(N)  
plt.xticks(ind, ('Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar'))
plt.legend((area_namerica['area_no'], area_namerica['area_yes'],area_namerica['area_maybe']), ('No', 'Yes', 'Maybe'))

plt.show()
plt.tight_layout()
plt.savefig('classified_area_casia.png',dpi=300)
fig.savefig(os.path.join(mypath,'classified_area_casia.png'))
```
